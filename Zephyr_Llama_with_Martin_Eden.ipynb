```python
# setup data
from llama_index import SimpleDirectoryReader

# load in some sample data
reader = SimpleDirectoryReader(input_files=["martineden.txt"])
documents = reader.load_data()
```

```python
from llama_index.llama_pack import download_llama_pack

# download and install dependencies
ZephyrQueryEnginePack = download_llama_pack(
  "ZephyrQueryEnginePack", "./zephyr_pack"
)
```

```python
# You can use any llama-hub loader to get documents!
zephyr_pack = ZephyrQueryEnginePack(documents)
```

```python
# run query
response = zephyr_pack.run("What did Martin Eden do for Ruth?", similarity_top_k=2)
print(str(response))
```

```python
# run query
response = zephyr_pack.run("Martin Eden personality", similarity_top_k=2)
print(str(response))
```

```python
# inspect modules in zephyr pack
print(zephyr_pack.get_modules().keys())
```

```python
# Use the index directly
index = zephyr_pack.index
query_engine = index.as_query_engine()
retriever = index.as_retriever(similarity_top_k=2)
```

```python
from llama_index.response.notebook_utils import display_source_node
results = retriever.retrieve("What did Martin Eden do for Ruth?")
for n in results:
  display_source_node(n, source_length=1000)
```

```python
from llama_index.response.notebook_utils import display_source_node
results = retriever.retrieve("Martin Eden personality")
for n in results:
  display_source_node(n, source_length=1000)
```
